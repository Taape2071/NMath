### a) ###

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gammaincinv, gammainccinv

def euler(x):
    return (1+x) ** (1/x)

empty = np.array([10E-19, 10E-16, 10E-13, 10E-10, 10E-7, 10E-4, 10E-1])

euler_number = np.zeros(7)

for i in range(7):
    euler_number[i] = np.exp(1)

euler_approx = euler(empty)

fig = plt.figure(figsize=(13,6))
ax = fig.add_subplot(111)
ax.plot(empty, euler_number)
ax.plot(empty, euler_approx)
plt.xlabel("h values")
plt.ylabel("e")
plt.legend(["Euler", "Approximation"])
ax.set_xscale('log')
plt.title("Euler number approximation")

plt.show()
```

If we evaluate (1 + h) ^ 1/h for smaller values of h, the result should get closer to e.
Yet, see for yourself what happens as h approaches 10^−16. The approximation starts to deviate from the value e, and we will
eventually get 1. This happens because h < 10^−16 is below machine accuracy, so
that 1 + h rounds up to 1.

Second order so we get O(3√epsilon) which is equal to 10^-5.3333

```python
epsilon = 10E-16

def u(x):
    return np.exp(x)

#u(x) = u'(x) because u(x) = e^x

def ForwardDiff(x,h):
    return ((u(x+h) - u(x))/h),  ((u(x+h) - u(x))/h) - u(x)

def CentralDiff(x,h):
    return ((u(x+h) - u(x-h))/(2*h)), ((u(x+h) - u(x-h))/(2*h)) - u(x)
 

newEmpty = np.array([10E-12, 10E-10, 10E-8, 10E-6, 10E-4, 10E-2, 1])

forwardDifferentiation, forwardError = ForwardDiff(0,newEmpty)
centralDifferentiation, centralError = CentralDiff(0,newEmpty)


fig = plt.figure(figsize=(13,6))
ax = fig.add_subplot(111)
ax.plot(newEmpty, forwardError)
ax.plot(newEmpty, centralError)
plt.xlabel("h values")
plt.ylabel("error and values")
plt.legend(["Forward Error", "Central Error"])
ax.set_xscale('log')
ax.set_yscale('log')
plt.title("Forward and Central Differentiation")
```
