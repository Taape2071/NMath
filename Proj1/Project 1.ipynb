```python
import numpy as np
import scipy.fft
import matplotlib.pyplot as plt

def transform(f, N, start=0.0):
    #Creating a set of equidistant points
    x = np.linspace(start, start + 1, N, endpoint=False)
    
    #Gives the values of f for each equidistant point
    fValues = f(x)
     
    return np.transpose(fValues)

#Each function
def f1(x):
    return np.sin(8 * np.pi * x)

def f2(x):
    return np.sin(32 * np.pi * x) + np.cos(128 * np.pi * x)

def f3(x):
    return x

def f4(x):
    return 1 - np.abs(x)

#Starting values
start = -0.5

Ns = np.array([5, 17, 257])

N_max = max(Ns)
x = np.linspace(start, start + 1, N_max, endpoint=False)

# Plotting the functions and their transforms for different values of N
for N in Ns:
    #Transforms
    f1_hat = scipy.fft.fft(transform(f1, N, start))
    f2_hat = scipy.fft.fft(transform(f2, N, start))
    f3_hat = scipy.fft.fft(transform(f3, N, start))
    f4_hat = scipy.fft.fft(transform(f4, N, start))

    #Plotting with real and imaginary values
    fig, axs = plt.subplots(2, 4, figsize=(16, 6))

    axs[0, 0].plot(x, f1(x).real, label='Real')
    axs[0, 0].plot(x, f1(x).imag, label='Imag')
    axs[0, 0].set_title(f'$f_1$ for N = {N}')
    axs[0, 0].legend()

    axs[0, 1].plot(x, f2(x).real, label='Real')
    axs[0, 1].plot(x, f2(x).imag, label='Imag')
    axs[0, 1].set_title(f'$f_2$ for N = {N}')
    axs[0, 1].legend()

    axs[0, 2].plot(x, f3(x).real, label='Real')
    axs[0, 2].plot(x, f3(x).imag, label='Imag')
    axs[0, 2].set_title(f'$f_3$ for N = {N}')
    axs[0, 2].legend()

    axs[0, 3].plot(x, f4(x).real, label='Real')
    axs[0, 3].plot(x, f4(x).imag, label='Imag')
    axs[0, 3].set_title(f'$f_4$ for N = {N}')
    axs[0, 3].legend()

    # Plot the real and imaginary parts of f_hat
    axs[1, 0].plot(np.arange(-N // 2, N // 2), f1_hat.real, label='Real')
    axs[1, 0].plot(np.arange(-N // 2, N // 2), f1_hat.imag, label='Imag')
    axs[1, 0].set_title(f'$\^f_1$ for N = {N}')
    axs[1, 0].legend()

    axs[1, 1].plot(np.arange(-N // 2, N // 2), f2_hat.real, label='Real')
    axs[1, 1].plot(np.arange(-N // 2, N // 2), f2_hat.imag, label='Imag')
    axs[1, 1].set_title(f'$\^f_2$ for N = {N}')
    axs[1, 1].legend()

    axs[1, 2].plot(np.arange(-N // 2, N // 2), f3_hat.real, label='Real')
    axs[1, 2].plot(np.arange(-N // 2, N // 2), f3_hat.imag, label='Imag')
    axs[1, 2].set_title(f'$\^f_3$ for N = {N}')
    axs[1, 2].legend()

    axs[1, 3].plot(np.arange(-N // 2, N // 2), f4_hat.real, label='Real')
    axs[1, 3].plot(np.arange(-N // 2, N // 2), f4_hat.imag, label='Imag')
    axs[1, 3].set_title(f'$\^f_4$ for N = {N}')
    axs[1, 3].legend()

    plt.tight_layout()
    plt.show()

```

#### $f_1$ and $f_2$ have the best approximations. We can see from the plots, that these two functions oscillate during the interval and the DFT and f can recreate these frequencies. On the other side, the graph of the transforms of $f_3$ and $f_4$ is different than the functions themselves and it does not create good approximations. This happens because $f_3$ is at most piecewise periodic and $f_4$ is "triangular", meaning that if we increase the sample value the approximation will be more adequate. ####

#### $c_k(f) = \frac{1}{N} \sum_{j = 0}^{N-1} f_j e^{\frac{-2 \pi i j k}{N}} \\$ In this case we have: $\\ c_k(D_n) = \frac{1}{N} \sum_{j = 0}^{N-1} D(\frac{j}{N}) e^{\frac{-2 \pi i j k}{N}} = \frac{1}{N} \sum_{j = 0}^{N-1} (1 + 2 \sum_{l = 1}^{n} \cos(\frac{2 \pi l j}{N}) ) e^{\frac{-2 \pi i j k}{N}} = \\ = \frac{1}{N} \sum_{j = 0}^{N-1}  e^{\frac{-2 \pi i j k}{N}} + \frac{2}{N} \sum_{j = 0}^{N-1} \sum_{l = 1}^{n} \cos(\frac{2 \pi l j}{N}) e^{\frac{-2 \pi i j k}{N}}\\$ From 1f) we can deduce that the first term of $c_k(D_n)$ will be 1 when $k \mod N \equiv 0$ and 0 for the rest of the k values. For the second term we have: $\\ \cos(\frac{2 \pi l j}{N}) = \frac{1}{2}(e^{\frac{2 \pi i l j }{N}} + e^{\frac{-2 \pi i l j}{N}}) \\ $ So we apply this to the expression and get: $\\ \frac{2}{N} \sum_{j = 0}^{N-1} \sum_{l = 1}^{n} \cos(\frac{2 \pi l j}{N}) e^{\frac{-2 \pi i j k}{N}} = \frac{1}{N} \sum_{j = 0}^{N-1} \sum_{l = 1}^{n}  (e^{\frac{2 \pi i l j }{N}} + e^{\frac{-2 \pi i l j}{N}})  e^{\frac{-2 \pi i j k}{N}} = \frac{1}{N} \sum_{j = 0}^{N-1} \sum_{l = 1}^{n}  (e^{\frac{-2 \pi i j }{N}(k-l)} + e^{\frac{-2 \pi i j}{N}(k + l)}) \\ $ We can now separate this expression in two sums, evaluating both expressions with what we have learned from 1f). So: $\\ \frac{1}{N} \sum_{j = 0}^{N-1} \sum_{l = 1}^{n}  e^{\frac{-2 \pi i j }{N}(k-l)} =\begin{cases}1 & \text{if } (k - l) \mod N \equiv 0, \\0 & \text{else}.\end{cases}\\$ $\\ \frac{1}{N} \sum_{j = 0}^{N-1} \sum_{l = 1}^{n}  e^{\frac{-2 \pi i j }{N}(k+l)} =\begin{cases}1 & \text{if } (k + l) \mod N \equiv 0, \\0 & \text{else}.\end{cases}\\$ This means that for all the values that are below N, $c_k(D_n) = 1$ if $k-l = 0$ or $k + l = 0$ or $k = 0$. We can then write $c_k(D_n)$ as follows: $c_k(D_n) = \begin{cases}1 & \text{if } -n \leq k \leq n, \\0 & \text{if } k > n.\end{cases}$   ####

## d) ##

```python
#Gets N and n and returns the fourier coefficients of Dirichlet kernel.
def dirichlet_ck(N, n, start=0.0):
    #x equidistant points
    x = np.linspace(start, start + 1, N, endpoint=False)

    #create c_k, only looking for positive frequencies
    ck = np.empty(N)

    for k in range(N):
        if k <= n:
            ck[k] = 1
        else:
           ck[k] =  0 
    
    return ck

#Starting values
N = 512
n = 48

#Convolutes two transforms
def conv(a,b):
    convolute = np.empty(len(a))
    for i in range(len(a)):
        convolute[i] = a[i] * b[i]
    
    return convolute

#Array with all the frequencies
frequency_array = np.empty(N)
for i in range(N):
    frequency_array[i] = i+1

#f_2 transform and ck(Dn)
f2_hat512 = scipy.fft.fft(transform(f2, N, start))
ck_48 = dirichlet_ck(N,n,start)

#Convolution and plotting
convolution = conv(f2_hat512,ck_48)

fig, ax = plt.subplots(figsize=(16, 6))
ax.plot(frequency_array, convolution)
ax.set_title(f'Convolution using Dirichlet kernel and $f_2$')
plt.xlabel('Frequency')
plt.ylabel('Signal')
ax.legend()
```

#### After convolution the signal is now filtered. In this case, the Dirichlet kernel applies a low-pass filter, meaning that it will reduce the high-frequency components and keep the low-frequency signals, making the original signal smoother. Mathematically we keep all the low-frequency values by multiplying them with 1 and reduce the high-frequency values by multiplying these with 0. ####
