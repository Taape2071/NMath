If we have no points, then $n=0$ and $i=0$. This falls out of reach of the recursive formula but $B_{0,0} = 1$ for all $0<t<1$. <br>

If $n=2$, $B_{0,2} =(1-t)^2$ and $B_{2,2}=t^2$, however $i=0$ or $i=n$ are also not valid. <br>    
  
Now we evaluate $B_{1,2} = 2t(1-t)$: <br>

$B_{1,2}(t) = (1 - t) B_{1, 1}(t) + t B_{0, 1}(t) = (1-t)t + t(1-t) = 2t(1-t)$ <br> <br>
In this case, the recursive formula is valid for the first case of its supposed reach. Now assume it holds for $n = k$, $0<i<k$: <br> <br>
$B_{i,n}(t) = \binom{n}{i} t^i (1-t)^{n-i} = B_{i,n}(t) = (1 - t) B_{i, n-1}(x) + t B_{i-1, n-1}(t)$

### d) ###

In this task we use the mathematical definition of a Bernstein polynomial shown at the start of the project and derive it. In addition, we use that: <br> <br>
$\binom{n}{i} = \frac{n!}{i! \cdot (n - i)!}$ <br> <br>
Now we can start deriving: <br> <br>
$\frac{d}{dt} B_{i,n}(t) = \frac{d}{dt} \left( \binom{n}{i} \cdot t^i \cdot (1 - t)^{n - i} \right)$ <br> <br>
Then we use the product rule and use the definition of the binomial: <br> <br>
$\frac{d}{dt} B_{i,n}(t) = \frac{in!}{i!(n-i)!}t^{i-1}(1-t)^{n-i} - \frac{(n-i)n!}{i!(n-i)!} t^i (1-t)^{n-i-1}$ <br>
$ = \frac{n(n-1)!}{(i-1)!(n-i)!}t^{i-1}(1-t)^{n-i} - \frac{n(n-1)!}{i!((n-1)-i)!} t^i (1-t)^{n-i-1}$ <br>
$ = n(\frac{(n-1)!}{(i-1)!(n-i)!}t^{i-1}(1-t)^{n-i} - \frac{(n-1)!}{i!((n-1)-i)!} t^i (1-t)^{n-i-1}) $ <br>
<br>
We can see that we have an expression for $B_{i-1,n-1}(t)$ on the left hand side of the minus sign, as well as an expression for $B_{i,n-1}(t)$ on the right hand side. So we conclude that: <br> <br>
$ \frac{d}{dt} B_{i,n}(t) = n(B_{i-1,n-1}(t) - B_{i,n-1}(t))$

```python
import numpy as np
import matplotlib.pyplot as plt

def de_Casteljau(P, t):# P is a vector– of n+1 points, evaluate the corresponding Bézier curve at t(returnvalue  bezier_point).
    n = len(P) - 1  # degree of the Bézier curve
    Pvecs = [P]  #  intermediate List, Pvecs that contains a vector of points for every “level” k considered.

    for k in range(1, n + 1):
        intermediate_points = []
        for i in range(n - k + 1):
            point = (1 - t) * Pvecs[k-1][i] + t * Pvecs[k-1][i + 1]    # intermediate points by de Casteljau formula
            intermediate_points.append(point) # collect all of one level
        Pvecs.append(intermediate_points) # add for each level

    # The final point : element of the last level
    bezier_point = Pvecs[-1][0]

    return bezier_point, Pvecs


# control points
p0 = np.array([0, 0])
p1 = np.array([-1, 5])
p2 = np.array([4, 4])
p3 = np.array([9, 3])
p4 = np.array([8, 8])
P = [p0, p1, p2, p3, p4]


t0 = 1/3 # given t for evaluation of Bezier curve
bezier_point, Pvecs = de_Casteljau(P, t0)

t_values = np.linspace(0, 1, 100)  # Generate a range of t values

curve_points = []
for t in t_values:
    bezier_point, _ = de_Casteljau(P, t)
    curve_points.append(bezier_point)

# Convert the list of points to a NumPy array
curve_points = np.array(curve_points)

# Plot the Bézier curve
plt.plot(curve_points[:, 0], curve_points[:, 1], label='Bézier Curve')

# Plot the Bézier curve and intermediate points
curve_points0 = np.array([p[0] for p in Pvecs])
plt.plot(curve_points0[:, 0], curve_points0[:, 1], label=f't={t0}')
for k in range(len(Pvecs)):
    points = np.array(Pvecs[k])
    plt.plot(points[:, 0], points[:, 1], '--o', label=f'Level {k}')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bézier Curve with Intermediate Points')
plt.grid(True)
plt.show()
```

```python
def b3(t, p0, p1, p2, p3):
    # Cubic Bézier curve
    return (
        (1 - t) ** 3 * p0 +
        3 * (1 - t) ** 2 * t * p1 +
        3 * (1 - t) * t ** 2 * p2 +
        t ** 3 * p3
    )

def composite_Bezier(P, t):
    m, n, _ = P.shape  # Get the number of segments and number of control points

    # Find the segment index i for the given t
    i = int(t)
    i = min(max(i, 0), m - 1)  # Ensure i is within the valid range

    # Calculate the local parameter within the segment
    local_t = t - i

    # Evaluate the Bézier curve for the selected segment
    p0, p1, p2, p3 = P[i]
    result = b3(local_t, p0, p1, p2, p3)

    return result

# Define the control points matrix
P = np.array([
    [(-1, 0), (-1, 1/3), (-1/3, 1), (0, 1)],
    [(0, 1), (1/3, 1), (1, 1/3), (1, 0)],
    [(1, 0), (1, -1/3), (1/3, -1), (0, -1)]
])

# Plot the composite curve
t_values = np.linspace(0, len(P), 1000)  # Generate values of t
curve_points = [composite_Bezier(P, t) for t in t_values]

# Extract x and y coordinates for plotting
x_values, y_values = zip(*curve_points)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="Cubic Composite Bézier Spline")
plt.scatter(P[:,:,0], P[:,:,1], c='red', marker='o', label="Control Points")
plt.title("Cubic Composite Bézier Spline")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

This plot gives us a non-periodic spline with 3 different curves, since it is a non-closed spline. We can also observe that we have $C^1$ continuity as the neighboring control points are symmetric.

  b3(t - i + 1; p_{0,i}, p_{1,i}, p_{2,i}, p_{3,i}) & \text{for } i - 1 \leq t < i \text{ and each } i = 0, \ldots, m -1 \\
  b3(t - m + 2; a_{m-1}, p_{1,m-1},2 a_0 - P_{1,1}, a_0) & \text{for i = m - 1} 
\end{cases}
$ <br> <br>

$B(i)=a_i$ simply means that we have $a_i$ values as endpoints for each cubic bezier curve we spline together.  <br> <br>
$B'(i) =v_i$ means that we need $P_{1,i}$ to be  given by $P_1 = a_i + \frac{v_i}{3}$, from $b_3 ' (0) = 3(P_1-P_0)$ for all i. <br> <br> 
But this means we also need $ P_2 = a_{i+1} - \frac{v_{i+1}} {3}$, from $b_3 ' (1) = 3(P_3-P_2) = 3(a_{i+1} -P_2) = v_{i+1}$ for all i. <br> <br>
$B(t) =
\begin{cases}
  b3(t - i + 1; a_i, a_i + \frac{v_i}{3},a_{i+1} - \frac{v_{i+1}}{3}  , a_{i+1}) & \text{for } i - 1 \leq t < i \text{ and each } i = 0, \ldots, m-1 \\
  b3(t - m + 2; a_{m-1}, p_{1,m-1}, 2a_0 - P_{1,1}, a_0) & \text{for i = m - 1} 
\end{cases}
$ <br> <br>

This periodic bezier curve is derived from the essential properties required to construct an algorithm, that were obtained from this task. We will now implement these properties in the following function.

```python
def interpolate_periodic(A, V):
    m = len(A)
    P = []

    # For i from 0 to m-1
    # Use the computed algorithm
    for i in range(m): 
        ai = A[i]
        vi = V[i]
        ai1 = A[(i + 1) % m]
        vi1 = V[(i + 1) % m]

        P0 = ai
        P3 = ai1
        P1 = ai + vi / 3   
        P2 = ai1 - vi1 / 3   

        P.append([P0, P1, P2, P3])
    
    # Treating i = m separately
    # Ensure periodicity , B(0)=B(m)
    P[0][0] = P[-1][3]      

    # Adjust P0 and P3 to match derivatives
    derivative_at_start = (P[1][0] - P[0][0]) * 3
    derivative_at_end = (P[-1][3] - P[-2][3]) * 3
    P[0][3] = P[0][0] + derivative_at_start / 3
    P[-1][0] = P[-1][3] - derivative_at_end / 3

    return np.array(P)
```

```python
# Define the number of rows and columns in P
num_matrices, num_rows, num_cols = P.shape

# Arrays for Q, R, S, and T
Q = np.zeros((num_matrices, num_rows, num_cols))
R = np.zeros((num_matrices, num_rows, num_cols))
S = np.zeros((num_matrices, num_rows, num_cols))
T = np.zeros((num_matrices, num_rows, num_cols))

mat_R = np.array([[np.cos(np.pi/8), -np.sin(np.pi/8)],
                    [np.sin(np.pi/8), np.cos(np.pi/8)]])
mat_S = np.array([[1, 1/2],
                    [1, 0]])
mat_T = np.array([[-1, -2/3],
                    [1, 0]])

# Transformations for each i and j
for i in range(num_matrices):
    for j in range(num_rows):
        # Apply the transformations element-wise
        Q[i, j, :] = 2 * P[i, j, :]
        R[i, j, :] = np.dot(mat_R, P[i, j, :])
        S[i, j, :] = np.dot(mat_S, P[i, j, :]) + (1/3) * np.array([1, 1])
        T[i, j, :] = np.dot(mat_T, P[i, j, :]) + (1/4) * np.array([1, -1])

t_values = np.linspace(0, len(R), 1000)

# Create subplots for all four plots
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
fig.suptitle("Periodic Composite Cubic Bézier Spline")

# Generate and plot curves for Q, R, S, and T
for i, data in enumerate([(Q, "Q"), (R, "R"), (S, "S"), (T, "T")]):
    points, label = data
    curve_points = [composite_Bezier(points, t) for t in t_values]
    x_values, y_values = zip(*curve_points)
    
    row, col = i // 2, i % 2
    axs[row, col].plot(x_values, y_values, label="Periodic Composite Cubic Bézier Spline")
    axs[row, col].set_title(label)
    axs[row, col].set_xlabel("X")
    axs[row, col].set_ylabel("Y")
    axs[row, col].legend()
    axs[row, col].grid(True)


plt.tight_layout()
plt.show()
```

The Q transformation scales the control points by a factor of 2. This means that each control point is multiplied by 2 in both the x and y directions. <br>
The effect of the Q transformation is to increase the magnitude of the control points, making the curve more "stretched out" in both dimensions. <br> <br>

The R transformation involves a rotation of the control points by an angle of π/8 (45 degrees in the counterclockwise direction).<br>
This rotation changes the orientation of the control points, potentially changing the direction and shape of the curve.<br> <br>

The S transformation consists of two operations: a scaling and a translation. <br>
It scales the control points by a factor of 1 in the x-direction and 1/2 in the y-direction. This has the effect of changing the ratio of the control points.<br>
It also translates the control points by (1/3, 1/3), moving them up and to the right by a certain amount. <br>
The combination of scaling and translation changes the shape and position of the curve. <br> <br>


The T transformation involves a transformation and a translation. <br>
It shears the control points by applying a transformation matrix that skews them along the x-axis. <br>
It also translates the control points by (-1/4, -1/4). <br>
The shearing and translation twists the shape of the curve, especially in the x-direction.
