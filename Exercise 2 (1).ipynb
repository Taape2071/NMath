```python
def cheb_nodes(N):
    validN = np.arange(1, N + 1)
    nodes = np.cos((2 * validN - 1) * np.pi / (2 * N))
    return nodes

def cheb_coef(y, t):
    N_here = len(t)
    c_0 = (np.sum(y) * 2) / N_here  

    c_k = np.zeros(N_here)
    for k in range(1, N_here):
        T_k = FirstDefChebyshev(k, t)
        c_k[k] = (2 / N_here) * np.sum(y * T_k)

    return c_0, c_k

def cheb_interpol(c, s):
    c_0, c_k = c
    N_here = len(c_k)
    inter = c_0 / 2  
    
    for k in range(1, N_here):
        T_k = FirstDefChebyshev(k,s)
        inter += c_k[k] * T_k
    
    return inter

# Example usage:
a = -2
b = 2
N1 = np.array([4,6,8,11])

fig = plt.figure(figsize=(13,6))
ax = fig.add_subplot(111)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Chebyshev Interpolation, First definition")

for i in N1:

    x = cheb_nodes(i)
    f_1 = x ** 11 - 3 * x ** 7 + 6
    t = (2 * x - a - b) / (b - a)
    c = cheb_coef(f_1, t)
    test_nodes = np.linspace(a, b, 100)
    inter_values = cheb_interpol(c, (2 * test_nodes - a - b) / (b - a))
    ax.plot(test_nodes, inter_values)
    ax.legend(["N = 4", " N = 6", "N = 8", "N = 11"])

```
